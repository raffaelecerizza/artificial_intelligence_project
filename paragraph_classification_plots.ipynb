```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mtl_train_loss = pd.read_csv("./plots/mtl_train_loss.csv")
mtl_val_loss = pd.read_csv("./plots/mtl_val_loss.csv")
```

```python
mtl_train_loss
```

```python
steps = mtl_train_loss['Step'].values
steps
```

```python
mtl_train_loss_values = mtl_train_loss['Value'].values
mtl_train_loss_values
```

```python
mtl_val_loss
```

```python
mtl_val_loss_values = mtl_val_loss['Value'].values
mtl_val_loss_values
```

```python
plt.figure(figsize=(10, 5))
plt.plot(steps, mtl_train_loss_values, label='Training Loss', marker='o', linestyle='-', color='b')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()
plt.show()
```

```python
mtl_val_seqc_f1 = pd.read_csv("./plots/mtl_val_seqc_f1.csv")
```

```python
steps = mtl_val_seqc_f1['Step'].values
steps
```

```python
epochs = list(range(1, 11))
epochs
```

```python
mtl_val_seqc_f1_values = mtl_val_seqc_f1['Value'].values
mtl_val_seqc_f1_values 
```

```python
plt.figure(figsize=(10, 5))
plt.plot(epochs, mtl_val_seqc_f1_values, label='Validation F1-score')
plt.xlabel('Epoch')
plt.ylabel('F1-score')
plt.title('Validation F1-score for the Multi-task model')
plt.legend()
plt.show()
```
